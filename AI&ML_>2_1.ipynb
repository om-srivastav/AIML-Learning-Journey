#LOOP over dict

```python
student
```

```python
for k in student.keys():
  print(k)#only prints the keys, not value
```

```python
for k in student.keys():
  print(student[k])#-->values get printed
```

```python
for v in student.values():
  print(v)
```

```python
for elem in student.items():
  print(elem)#-->gives tuple of key-value pair
```

```python
for elem in student.items():
  print("key =",elem[0], ", value=",elem[1])
```
