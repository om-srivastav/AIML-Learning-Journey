```python
#elif-->else-if
# marks =>
# marks < 30 --> "Low"
# marks b/w 31 and 70 --> "Avg"
# marks > 70 --> "High"
# using if-else only
marks=int(input("Enter marks "))
if marks < 30:
  print("Low")
else:
  if(marks >=31) and (marks <= 70):
    print("Avg")
  else:
   print("High")
```

```python
# by elif--> for many conditions
# else if => elif-->jb true tb vhi se breakkkk
marks=int(input("Enter the marks"))
if marks<30:
  print("Low")
elif (marks >=31) and (marks <= 70):
  print("Avg")
else:
  print("High")
```
