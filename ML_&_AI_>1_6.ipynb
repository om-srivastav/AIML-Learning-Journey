```python
for i in range(1,6,1):#-->range(start,end,step)
  print(i)
```

```python
#Print all the numbers from 2 to 10(both inclusive)
for i in range(2,11,1):
  print(i,end = " ")
  #we use 'end='_'' for preventing new line
```

```python
#Print 10 to 1
for i in range(10,0,-1):
  print(i)
#for negative decrement if end is n then the loop will terminate at n+1
```

```python
# print 10 to 1 only by 1 parameter
for i in range(10):
  print(10-i)
```
