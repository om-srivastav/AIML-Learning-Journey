#Break and Continue

```python
for i in range(1,11):
  if i==5:
    break#loop will terminate at i =5, loop will run to 4
  print(i)
```

```python
for i in range(1,11):

  print(i)

  if i==5:
    break#loop will terminate at i =5, here loop will run to 5
```

#continue statement

```python
for i in range(1,10):

  if i==5:
    break
  print(i)
```

```python
for i in range(1,10):

  if i==5:
   continue #as when i=5, it will skipped
  print(i)
```

```python
for i in range(1,10):
  print(i)
  if i==5:
   continue #not executed
```

```python
#print all even number b/w 1 to 10---continue statement
for i in range(1,11):
  if i%2 !=0:
    continue
  print(i)
```
