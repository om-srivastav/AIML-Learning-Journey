```python
'''body mass index (bmi) is a simple measure that helps determine  whether a person's weight is appropriate for their height
wap that takes bmi as input and displays the corresponding weight category using rules--->
bmi value category:
less than 18.5 --> underweight
18.5 to 24.9 -->normal
25.0 to 29.9 -->over
30.0 or above obese
{bmi = wt(kg)/h(m)^2}'''
height=float(input("enter ur height in cm "))
height/=100.0
wt=float(input("enter ur weight in kg "))
BMI=wt/(height * height)
print("YOUR BMI IS-->",BMI)
if BMI < 18.5:
  print("underwight")
elif BMI <=24.9:
  print("normal weight")
elif BMI <=29.9:
  print("overweight")
else:
  print("obese")
```
